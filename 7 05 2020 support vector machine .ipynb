```python
import pandas as pd
import numpy as np
from collections import Counter
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.svm import SVC
```

```python
data=pd.read_csv('C:/Users/Admin/Desktop/Kittu/car_evaluation.csv')
data.head()
```

```python
data.shape
```

```python
X=data.iloc[:,:-1]
y=data.outcome
```

```python
enc=LabelEncoder()
X.buying=enc.fit_transform(X.buying)
X.maint=enc.fit_transform(X.maint)
X.lug_boot=enc.fit_transform(X.lug_boot)
X.safety=enc.fit_transform(X.safety)
X.head(3)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=10)
model=SVC(C=100,gamma=0.1,kernel='rbf') # C 0.1 to 1000, gamma 0.01 to 10
model.fit(X_train,y_train)#Training the model
```

```python
y_predict=model.predict(X_test)
accuracy_score(y_test,y_predict)
```

```python
print(Counter(y_test))
pd.crosstab(y_test,y_predict)
```

Grid Search Cross Validation

```python
from sklearn.model_selection import GridSearchCV
parameters = {'kernel': ['rbf','linear','poly','sigmoid'],'C': [1,10,100,500],'gamma': [0.01,0.1,0.5,1.2]}
grid_model = GridSearchCV(SVC(random_state=10),parameters,verbose=2)
grid_model.fit(X_train,y_train)
```
